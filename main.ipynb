$$\text{\underline{Pilot to full analysis}}$$

$\text{Defining Braggs Law:} $

$$ 2{d}\sin{\theta{}} = n\lambda{} $$

$\text{In our case:} $

$$\lambda{}_{\alpha{}} = 0.7108 \AA $$
$$\lambda{}_{\beta{}} =  0.6309 \AA $$

$$ 2.5^{\circ{}} < \theta{} < 31^{\circ{}} $$

$ \text{Literature values for d at room temperature} $

$$ d_{NaCl} = 2.814 \AA $$
$$ d_{Si} = 5.430 \AA $$

$\text{Using this information we can say that where braggs law is true peaks must arise, essentially the reverse of the experiment i.e. :}$


```python

def find_peaks(n,wavelength,d):
    angle_range = np.linspace(2.5,30,276)
    array_to_find = abs((2*d*np.sin(angle_range/180 * pi)) - n*wavelength)
    desired = list(array_to_find).index(min(array_to_find))

    return((desired))


nacl_peaks = []
si_peaks = []


for n in range(1,10):
    for wavelength in [0.7108,0.6309]:
        nacl_peaks.append(find_peaks(n,wavelength,2.814))
        si_peaks.append(find_peaks(n,wavelength,5.430/sqrt(3)))

while 275 in nacl_peaks:
    nacl_peaks.remove(275)
while 275 in si_peaks:
    si_peaks.remove(275)


print(nacl_peaks)
print(si_peaks)

```

```python
def find_nth_occurrence(l, value, n):
    indices = [index for index, item in enumerate(l)
               if item == value]

    if len(indices) < n:
        return -1

    return indices[n - 1]

def find_max(x_range,xvals,Ivals):
    intensity_range = []
    num_points = int((x_range[1]-x_range[0])/0.1) + 1
    x_range = np.linspace(x_range[0],x_range[1],num_points)
    for x in x_range:
        x = round(x,1)
        intensity_range.append(Ivals[list(xvals).index(x)])

    
    if xvals[list(Ivals).index(max(intensity_range))] in x_range:
        return xvals[list(Ivals).index(max(intensity_range))],max(intensity_range)
    else:
        return xvals[find_nth_occurrence(Ivals,max(intensity_range),3)],max(intensity_range)

peak_1_angle,peak_1_I = find_max((5.1,6.5),angle_range,NaCl_I_exp)
peak_2_angle,peak_2_I = find_max((2.5,10.0),angle_range,NaCl_I_exp)
peak_3_angle,peak_3_I = find_max((12.0,13.0),angle_range,NaCl_I_exp)
peak_4_angle,peak_4_I = find_max((13.0,15.0),angle_range,NaCl_I_exp)
peak_5_angle,peak_5_I = find_max((18.0,20.0),angle_range,NaCl_I_exp)
peak_6_angle,peak_6_I = find_max((20.0,25.0),angle_range,NaCl_I_exp)

print('n=1')
print(peak_1_angle,peak_1_I)
print(peak_2_angle,peak_2_I)
print('n=2')
print(peak_3_angle,peak_3_I)
print(peak_4_angle,peak_4_I)
print('n=3')
print(peak_5_angle,peak_5_I)
print(peak_6_angle,peak_6_I)
```

$$ \text{Bremsstrahlung Continuum} $$

```python
def kramas_law(xvals,lambda_min,Brem_peak,lat_space):
    yvals = []
    K = 4*Brem_peak*lambda_min*lambda_min
    print(K)
    for x in xvals:
        wave_l = 2 * lat_space * sin(x*(pi/180))
        y = K * ((wave_l)/(lambda_min) - 1  ) * (1 / (wave_l ** 2))
        yvals.append(y)

    return yvals

```

```python
def kramas_law2(xvals,lambda_min,Brem_peak,lat_space):
    yvals = []
    M = 1.1*(10**-11)

    K = 4*Brem_peak*M*M
    print(K)
    for x in xvals:
        wave_l = 2 * lat_space * sin(x*(pi/180))
        y = K * ((wave_l - (lambda_min-M))/(M) - 1  ) * (1 / ((wave_l-(lambda_min-M)) ** 2))
        yvals.append(y)

    return yvals
```

```python
figNa, axNa = plt.subplots()

peak_brem_angle_Na,peak_brem_I_Na = find_max((2.6,5.0),angle_range,NaCl_I_exp)


brem_theo_y_na = kramas_law(angle_range,(2.95*(10**-11)),peak_brem_I_Na,(2.834*(10**-10)))
brem_theo_y_na2 = kramas_law2(angle_range,(2.95*(10**-11)),peak_brem_I_Na,(2.834*(10**-10)))

axNa.set(title=" NaCl ~Using literature lattice constant",xlabel='Angle(°)',ylabel='Count Rate')

axNa.set(ylim=(0,3700))
axNa.plot(angle_range,NaCl_I_exp,linewidth=0.5)
axNa.plot(angle_range,brem_theo_y_na,linewidth=0.5,label='intial')
axNa.plot(angle_range,brem_theo_y_na2,linewidth=0.5,label='altered')

axNa.legend()




```
